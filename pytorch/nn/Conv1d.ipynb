NOTE: 一维卷积不代表卷积核只有一维，也不代表被卷积的feature是一维的。一维的意思是卷积的方向是一维的!!!

```python
m = nn.Conv1d(16, 33, 3, stride=2)
input = torch.randn(20, 16, 50)
output = m(input)
```

```python
output.size()
```
