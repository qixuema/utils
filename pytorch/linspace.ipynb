```python
import torch
```

```python
step = 8
a1 = torch.linspace(-1., 1., steps=step, dtype=torch.float) # 生成 [-1, 1] 区间内，8个数的等差数列
print(a1)
a2 = a1.view(1, step)
print(a2)
a3 = a2.expand(step, step)
print(a3)
a4 = a3.reshape(1, -1)
print(a4)
```

```python
step = 8
b1 = torch.linspace(-1., 1., steps=step, dtype=torch.float) # 生成 [-1, 1] 区间内，8个数的等差数列
print(b1)
b2 = b1.view(step, 1)
print(b2)
b3 = b2.expand(step, step)
print(b3)
b4 = b3.reshape(1, -1)
print(b4)
```

```python
a3.permute(-1, -2)==b3
```

```python
c = torch.cat([a4, b4], dim=0)
print(c)
print(c.size())
```
