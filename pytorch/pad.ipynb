```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python

beta_1 = 1e-4
beta_T = 0.02
T = 1000
betas = torch.linspace(beta_1, beta_T, T).double()
alphas = 1. - betas
alphas_bar = torch.cumprod(alphas, dim=0) # 对第dim 维度进行连乘
alphas_bar_prev = F.pad(alphas_bar, [1, 0], value=1)[:T] # F.pad: tensor扩充函数
```
